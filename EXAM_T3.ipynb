```python
import pandas as pd  
import matplotlib.pyplot as plt  
  
data = {  
    'Student Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],  
    'Attendance (%)': [90, 75, 85, 60, 95],  
    'Mark': [88, 70, 80, 55, 95]}  

df = pd.DataFrame(data)  

print(df)  

# Plotting the correlation  
plt.figure(figsize=(8, 5))  
plt.scatter(df['Attendance (%)'], df['Mark'], color='blue')  
 
plt.title('Correlation between Attendance and Marks')  
plt.xlabel('Attendance (%)')  
plt.ylabel('Mark (out of 100)')  
plt.grid()  

# Show the plot  
plt.show()
```
