1. Time Complexity Analysis

```python
#1.1

for i in range(n): 
    i *= i  
    if i >= n:  
        break
        
```

```python
#1.2
for i in range(n):  
    i = pow(2, i)  
    if i > (n + 1):  
        break
```

```python
#1.3
def complex_method(n):  
    for i in range(n):  
        print(i, n)  
        if n > 1:  
            complex_method(n - 1)  
        else:  
            return True
```

2. Recursion

2.1 Sum of Digits

```python
def sum_of_digits(n):  
    if n == 0:  
        return 0  
    return (n % 10) + sum_of_digits(n // 10)  
 
result = sum_of_digits(53872)  
print(result)
```

2.2 Sum of All Numbers Up to n

```python
def sum_upto(n):  
    if n == 0:  
        return 0  
    return n + sum_upto(n - 1)  

  
result = sum_upto(500) 
print(result)
```

2.3 Fibonacci Series

```python
def fibonacci(n):  
    if n == 0:  
        return 0  
    elif n == 1:  
        return 1  
    else:  
        return fibonacci(n - 1) + fibonacci(n - 2)  


result = fibonacci(5)
print(result)
```
