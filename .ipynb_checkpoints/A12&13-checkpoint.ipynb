```python
import requests  
from bs4 import BeautifulSoup  

url = 'https://gsom.spbu.ru/en/'  

resp = requests.get(url)
if resp.status_code == 200:  

    soup = BeautifulSoup(resp.content, "html.parser")  
    # List to hold PNG links  
    png_links = []

    # Finding all <img> tags and check their 'src' attribute for PNG links
    for img in soup.find_all("img", src=True):
        src = img['src']  
# Checking if the 'src' ends with ".png"
        if src.lower().endswith(".png"):  
# Append the found PNG link   
            png_links.append(src)

    # Prepend base URL if needed (handling relative links)
    base_url = "https://gsom.spbu.ru"  
    full_png_links = []  

    for link in png_links:  
        # If the link is relative, prepend the base URL  
        if link.lower().endswith(".png"):  
            full_png_links.append(base_url + link)
    count_png=len(full_png_links)

    # Print the extracted PNG links 
    print("Total_png_links: ", count_png)
    print("Extracted PNG Links:")  
    for link in full_png_links:  
        print(link) 
```

```python
def bubble_sort(arr):  
    n = len(arr)  
    # Traverse through all elements in the array  
    for i in range(n):  
        # Last i elements are already in place  
        for j in range(0, n - i - 1):  
            # Compare and swap if necessary  
            if arr[j] > arr[j + 1]:  
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  
    return arr  

# Testing the sorting function with different lists  
short_list = [64, 34, 25, 12, 22]  
long_list = [64, 34, 25, 12, 22, 11, 90, 53, 75, 88, 9, 1, 3, 7, 100]  

sorted_short_list = bubble_sort(short_list.copy())  
sorted_long_list = bubble_sort(long_list.copy())  

print("Sorted Short List:", sorted_short_list)  
print("Sorted Long List:", sorted_long_list)
```

```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
import statistics as st 

# Setting up the financial data  
# Income for each month  
Inc = np.array([2000, 1500, 3001, 1982, 7320, 2003, 3001, 1482, 7320, 3289])  
Exp = np.array([200, 150, 300, 198, 732, 200, 300, 198, 732, 200]) # Expenses for each month  
Inc_no_tax = Inc * 0.7 
saving = Inc - Exp # Calculating savings as the difference between income and expenses    
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]  

# Creating a DataFrame to organize the financial data  
data = {  
    "Income": Inc,   
    "Expenses": Exp,   
    "Saving": saving,   
    "Income_without_tax": Inc_no_tax  
}  
df1 = pd.DataFrame(data, index=months)  # DataFrame with months as index  

# Plotting the financial data  
plt.figure(figsize=(12, 8))  # Set figure size for the plots  

# Monthly Income Plot  
plt.subplot(3, 1, 1)  # Create the first subplot in a 3x1 grid  
plt.plot(months, Inc, color='blue', marker='o')  # Line plot for income  
plt.title('Monthly Income')  # Title for the income plot  
plt.xlabel('Months')  # Label for x-axis  
plt.ylabel('Income')  # Label for y-axis  
plt.grid()  # Add grid lines for better readability  

# Monthly Savings Bar Chart  
plt.subplot(3, 1, 2)  # Create the second subplot  
plt.bar(months, saving, color='orange')  # Bar chart for savings  
plt.title('Monthly Savings')  # Title for the savings plot  
plt.xlabel('Months')  # Label for x-axis  
plt.ylabel('Savings')  # Label for y-axis  
plt.grid()  # Add grid lines for better readability  

# Savings Distribution Pie Chart  
total_saving = saving.sum()  # Calculate total savings for pie chart (not used directly)  
plt.subplot(3, 1, 3)  # Create the third subplot  
plt.pie(saving, labels=months, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)  
plt.title('Savings Distribution by Month') # Pie chart showing the savings distribution by month 

# Adjust layout and show the plots  
plt.tight_layout() 
plt.show()  

# Calculating Average Income by Quarters  
quarters = {  
    'Quarter 1': Inc[:3],  # Income for Jan, Feb, Mar  
    'Quarter 2': Inc[3:6],  # Income for Apr, May, Jun  
    'Quarter 3': Inc[6:9],  # Income for Jul, Aug, Sep  
    'Quarter 4': Inc[9:]    # Income for Oct  
}  

# Calculating average income for each quarter  
average_income = {quarter: st.mean(income) for quarter, income in quarters.items()}  

# Printing the average income for each quarter  
print("Average Income per Quarter:")  
for quarter, avg in average_income.items():  
    print(f"{quarter}: {avg:.2f}")  # Print average income, formatted to two decimal places
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  

# Load the dataset  
car_c = sns.load_dataset("car_crashes")


plt.figure(figsize=(10, 6))  
sns.scatterplot(data=car_c, x='speeding', y='alcohol', hue='total', size='ins_losses', sizes=(20, 200), alpha=0.6)  
plt.title('Speeding vs. Alcohol-Impaired Driving')  
plt.xlabel('Percentage of Drivers Speeding')  
plt.ylabel('Percentage of Drivers Alcohol-Impaired')  
plt.legend(title='Total Drivers', bbox_to_anchor=(1, 1))  
plt.grid()  
plt.show()




plt.figure(figsize=(14, 8))  
sns.barplot(data=car_c, x='total', y='ins_premium', estimator='mean', palette='viridis')  
plt.title('Average Car Insurance Premium by Number of Drivers Involved in Fatal Collisions')  
plt.xlabel('Total Drivers Involved in Fatal Collisions (per billion miles)')  
plt.ylabel('Average Insurance Premium')  
plt.xticks(rotation=45)  
plt.grid()  
plt.show()


plt.figure(figsize=(10, 8))  
correlation_matrix = car_c.corr()  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)  
plt.title('Correlation Heatmap of Car Crashes Dataset')  
plt.show()


plt.figure(figsize=(10, 6))  
sns.boxplot(data=car_c, x='alcohol', y='ins_losses', palette='Set2')  
plt.title('Distribution of Insurance Losses by Alcohol Consumption')  
plt.xlabel('Percentage of Drivers Alcohol-Impaired')  
plt.ylabel('Insurance Losses per Insured Driver')  
plt.grid()  
plt.show()
```
