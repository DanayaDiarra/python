1. For the given ticket_no 0005432312164, find: who and where flew.

```python

%%sql $CONNECT_DATA  

SELECT T.passenger_name, A.city  
FROM Tickets T  
JOIN Ticket_flights TF USING(ticket_no)  
JOIN Flights F USING(flight_id)  
JOIN airports A ON F.arrival_airport = A.airport_code  
WHERE T.ticket_no = '0005432312164'
```

```python
%%sql $CONNECT_DATA 
SELECT DISTINCT passenger_name, city
FROM Tickets,airports
WHERE ticket_no = '0005432312164'
```

```python
%%sql $CONNECT_DATA 
SELECT   
    (SELECT passenger_name FROM Tickets WHERE ticket_no = '0005432312164' LIMIT 1) AS passenger_name,  
    (SELECT city   
     FROM airports   
     WHERE airport_code IN (  
         SELECT arrival_airport   
         FROM Flights   
         WHERE flight_id IN (  
             SELECT flight_id   
             FROM Ticket_flights   
             WHERE ticket_no = '0005432312164'  
         )  
     ) LIMIT 1) AS city
```

2. Extend that solution about Mrs. 'ELLA DMITRIEVA' - from where and to where she flew.

```python
%%sql $CONNECT_DATA
SELECT DISTINCT   
    departure_airport.city AS departure_city,  
    arrival_airport.city AS arrival_city  
FROM   
    Tickets t  
JOIN Ticket_flights tf ON t.ticket_no = tf.ticket_no  
JOIN Flights f ON tf.flight_id = f.flight_id  
JOIN airports arrival_airport ON f.arrival_airport = arrival_airport.airport_code  
JOIN airports departure_airport ON f.departure_airport = departure_airport.airport_code  
WHERE t.passenger_name = 'ELLA DMITRIEVA'
```
